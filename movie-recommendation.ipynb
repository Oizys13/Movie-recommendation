```python
#we merge the datasets on the id
credits = credits.rename(columns={'movie_id': 'id'})
data = pd.merge(credits, movies, on='id')
data.drop('title_y', axis=1, inplace=True)
data.drop('budget', axis=1, inplace=True)
data = data.rename(columns={'title_x':'title'})
data.head(3)
```

```python
movies_set=data[['id','title','overview','runtime','genres','keywords','cast','crew']]
movies_set.head()
```

```python
movies_set.isnull().sum()
```

```python
movies_set.dropna(inplace=True)
```

```python
movies_set.isnull().sum()
```

```python
movies_set.duplicated().sum()
```

```python
import ast

def deStringify(text):
    items= []
    data_list = ast.literal_eval(text)
    for item in data_list:
        items.append(item['name'])
    return items    
```

```python

movies_set['genres'] = movies_set['genres'].apply(deStringify)
movies_set['keywords'] = movies_set['keywords'].apply(deStringify)
```

```python
def deStringify_cast(text):
    items= []
    counter = 0
    data_list = ast.literal_eval(text)
    for item in data_list:
        if counter <3 :
            items.append(item['name'])
        counter+=1
    return items    
```

```python
movies_set['cast'] = movies_set['cast'].apply(deStringify_cast)
```

```python
movies_set.head()
```

```python
def deStringify_crew(text):
    items= []
    data_list = ast.literal_eval(text)
    for item in data_list:
        if item['job'] == 'Director':
            items.append(item['name'])
            break
    return items 
```

```python
movies_set['crew'] = movies_set['crew'].apply(deStringify_crew)
```

```python
movies_set.head()
```

```python
movies_set['overview'] = movies_set['overview'].apply(lambda text: text.split()) 
```

```python
movies_set.head(2)
```

```python
def stripping_spaces (text):
    items = []
    for item in text:
        items.append(item.replace(" ",""))
    return items    
```

```python

movies_set['cast']= movies_set['cast'].apply(stripping_spaces)
movies_set['crew']= movies_set['crew'].apply(stripping_spaces)
```

```python
movies_set['genres']= movies_set['genres'].apply(stripping_spaces)
movies_set['keywords']= movies_set['keywords'].apply(stripping_spaces)
```

```python
movies_set.head(5)
```

```python
movies_set['tags'] = movies_set['overview'] + movies_set['genres'] + movies_set['keywords'] + movies_set['cast'] + movies_set['crew']
```

```python
movies_set.head()
```

```python
df = movies_set[['id','title','tags']]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda text: " ".join(text))
df.head(3)
```

```python
df['tags'] = df['tags'].apply(lambda text: text.lower())
df.head(3)
```

The Porter Stemmer is a popular algorithm for stemming words, reducing them to their root form

```python
import nltk
from nltk.stem import PorterStemmer
```

```python
porter_stemmer = PorterStemmer()
```

```python
def stems(text):
    stems = []
    for word in text.split():
        stems.append(porter_stemmer.stem(word))
    return " ".join(stems)    
```

```python
df['tags'] = df['tags'].apply(stems)
```

CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vector = vectorizer.fit_transform(df['tags']).toarray()
```

```python
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(vector)
cosine_sim
```

```python
cosine_sim_df = pd.DataFrame(cosine_sim)
cosine_sim_df
```

```python
df[df['title'] == 'Spider-Man'].index[0]
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    sorted_distances = sorted(list(enumerate(cosine_sim[movie_index])), reverse= True, key= lambda x:x[1])
    recommended_movies = []
    for i in sorted_distances[1:11]:
        recommended_movies.append(df.iloc[i[0]].title)
    return recommended_movies
```

```python
recommend('Spider-Man')
```
